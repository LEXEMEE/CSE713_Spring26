```python
# =========================================================
# 1. INSTALL REQUIRED PACKAGES
# =========================================================
!pip -q install kagglehub imagehash opencv-python seaborn


# =========================================================
# 2. DOWNLOAD DATASET FROM KAGGLE
# =========================================================
import kagglehub

path = kagglehub.dataset_download("maxwellbernard/brain-tumor-mri-multi-class-dataset")
print("Path to dataset files:", path)


# =========================================================
# 3. INSPECT DATASET STRUCTURE
# =========================================================
import os

for root, dirs, files in os.walk(path):
    print(root)
    if len(dirs) == 0:
        break


# =========================================================
# 4. IMPORT LIBRARIES
# =========================================================
import os
import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
import imagehash

from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from sklearn.metrics import classification_report, confusion_matrix

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.applications.efficientnet import preprocess_input


# =========================================================
# 5. SET DATASET ROOT
# =========================================================
DATA_ROOT = path
print("Dataset root:", DATA_ROOT)


# =========================================================
# 6. DEFINE VALID IMAGE FORMATS AND LABEL ALIASES
# =========================================================
IMAGE_EXTS = {".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff"}

LABEL_ALIASES = {
    "glioma": ["glioma"],
    "meningioma": ["meningioma"],
    "pituitary": ["pituitary"],
    "no_tumor": ["no_tumor", "notumor", "no tumor", "no-tumor", "normal", "healthy"]
}


# =========================================================
# 7. FUNCTION TO INFER LABEL FROM FILE PATH
# =========================================================
def infer_label_from_path(path):
    path_lower = path.lower().replace("\\", "/")
    for label, aliases in LABEL_ALIASES.items():
        for alias in aliases:
            if alias in path_lower:
                return label
    return None


# =========================================================
# 8. SCAN DATASET AND BUILD DATAFRAME
# =========================================================
rows = []

for root, dirs, files in os.walk(DATA_ROOT):
    for fname in files:
        ext = os.path.splitext(fname)[1].lower()
        if ext in IMAGE_EXTS:
            fpath = os.path.join(root, fname)
            label = infer_label_from_path(fpath)
            if label is not None:
                rows.append({
                    "filepath": fpath,
                    "label": label
                })

df = pd.DataFrame(rows)

print("Total images found:", len(df))
print(df["label"].value_counts())
print(df.head())


# =========================================================
# 9. PLOT ORIGINAL CLASS DISTRIBUTION
# =========================================================
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x="label", order=df["label"].value_counts().index)
plt.title("Original Class Distribution")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.xticks(rotation=15)
plt.show()


# =========================================================
# 10. SHOW SAMPLE IMAGES
# =========================================================
sample_df = df.sample(min(12, len(df)), random_state=42).reset_index(drop=True)

plt.figure(figsize=(14, 10))
for i, row in sample_df.iterrows():
    img = cv2.imread(row["filepath"])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.subplot(3, 4, i + 1)
    plt.imshow(img)
    plt.title(row["label"])
    plt.axis("off")

plt.tight_layout()
plt.show()


# =========================================================
# 11. COMPUTE PERCEPTUAL HASH FOR DUPLICATE DETECTION
# =========================================================
def compute_phash(path):
    try:
        img = Image.open(path).convert("RGB")
        return str(imagehash.phash(img))
    except Exception:
        return None


# =========================================================
# 12. REMOVE UNREADABLE FILES AND EXACT DUPLICATES
# =========================================================
df["phash"] = df["filepath"].apply(compute_phash)
df = df.dropna(subset=["phash"]).reset_index(drop=True)

df_dedup = df.drop_duplicates(subset=["phash"]).reset_index(drop=True)

print("Before deduplication:", len(df))
print("After deduplication :", len(df_dedup))
print("Removed duplicates  :", len(df) - len(df_dedup))
print()
print(df_dedup["label"].value_counts())


# =========================================================
# 13. PLOT CLASS DISTRIBUTION AFTER CLEANING
# =========================================================
plt.figure(figsize=(8, 5))
sns.countplot(data=df_dedup, x="label", order=df_dedup["label"].value_counts().index)
plt.title("Class Distribution After Deduplication")
plt.xlabel("Class")
plt.ylabel("Number of Images")
plt.xticks(rotation=15)
plt.show()


# =========================================================
# 14. TRAIN / VALIDATION / TEST SPLIT
# =========================================================
train_df, temp_df = train_test_split(
    df_dedup,
    test_size=0.30,
    random_state=42,
    stratify=df_dedup["label"]
)

val_df, test_df = train_test_split(
    temp_df,
    test_size=0.50,
    random_state=42,
    stratify=temp_df["label"]
)

print("Train size:", len(train_df))
print("Validation size:", len(val_df))
print("Test size:", len(test_df))


# =========================================================
# 15. VISUALIZE SPLIT DISTRIBUTION
# =========================================================
split_df = pd.concat([
    train_df.assign(split="train"),
    val_df.assign(split="validation"),
    test_df.assign(split="test")
])

plt.figure(figsize=(10, 5))
sns.countplot(data=split_df, x="label", hue="split")
plt.title("Train / Validation / Test Distribution")
plt.xlabel("Class")
plt.ylabel("Count")
plt.xticks(rotation=15)
plt.show()


# =========================================================
# 16. CLAHE PREPROCESSING
# =========================================================
def clahe_rgb(img_rgb):
    lab = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(lab)

    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
    l2 = clahe.apply(l)

    lab2 = cv2.merge((l2, a, b))
    return cv2.cvtColor(lab2, cv2.COLOR_LAB2RGB)


# =========================================================
# 17. COMPARE ORIGINAL AND CLAHE IMAGE
# =========================================================
example_row = train_df.sample(1, random_state=1).iloc[0]

img_bgr = cv2.imread(example_row["filepath"])
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
img_clahe = clahe_rgb(img_rgb)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.imshow(img_rgb)
plt.title("Original")
plt.axis("off")

plt.subplot(1, 2, 2)
plt.imshow(img_clahe)
plt.title("After CLAHE")
plt.axis("off")

plt.tight_layout()
plt.show()


# =========================================================
# 18. DEFINE CLASS MAPPINGS
# =========================================================
CLASS_NAMES = ["glioma", "meningioma", "pituitary", "no_tumor"]
class_to_idx = {name: i for i, name in enumerate(CLASS_NAMES)}
idx_to_class = {i: name for name, i in class_to_idx.items()}

print(class_to_idx)


# =========================================================
# 19. SET PIPELINE PARAMETERS
# =========================================================
IMG_SIZE = 224
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE


# =========================================================
# 20. LOAD AND PREPROCESS FUNCTION
# =========================================================
def load_and_preprocess(path, label):
    path = path.numpy().decode("utf-8")
    label = label.numpy().decode("utf-8")

    img = cv2.imread(path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = clahe_rgb(img)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE))
    img = img.astype("float32")
    img = preprocess_input(img)

    label_idx = np.int32(class_to_idx[label])
    return img, label_idx


# =========================================================
# 21. TENSORFLOW WRAPPER
# =========================================================
def tf_wrapper(path, label):
    img, label_idx = tf.py_function(
        func=load_and_preprocess,
        inp=[path, label],
        Tout=[tf.float32, tf.int32]
    )
    img.set_shape((IMG_SIZE, IMG_SIZE, 3))
    label_idx.set_shape(())
    return img, label_idx


# =========================================================
# 22. CREATE TENSORFLOW DATASETS
# =========================================================
def make_dataset(df_part, training=False):
    ds = tf.data.Dataset.from_tensor_slices((
        df_part["filepath"].values,
        df_part["label"].values
    ))
    ds = ds.map(tf_wrapper, num_parallel_calls=AUTOTUNE)
    if training:
        ds = ds.shuffle(1024, reshuffle_each_iteration=True)
    ds = ds.batch(BATCH_SIZE).prefetch(AUTOTUNE)
    return ds

train_ds = make_dataset(train_df, training=True)
val_ds = make_dataset(val_df, training=False)
test_ds = make_dataset(test_df, training=False)


# =========================================================
# 23. SHOW PROCESSED IMAGES
# =========================================================
images, labels = next(iter(train_ds))

plt.figure(figsize=(14, 10))
for i in range(min(12, len(images))):
    img_to_show = images[i].numpy().copy()
    img_to_show = img_to_show - img_to_show.min()
    img_to_show = img_to_show / (img_to_show.max() + 1e-8)

    plt.subplot(3, 4, i + 1)
    plt.imshow(img_to_show)
    plt.title(idx_to_class[int(labels[i].numpy())])
    plt.axis("off")

plt.tight_layout()
plt.show()


# =========================================================
# 24. COMPUTE CLASS WEIGHTS
# =========================================================
y_train = train_df["label"].map(class_to_idx).values

class_weights_values = compute_class_weight(
    class_weight="balanced",
    classes=np.array(list(range(len(CLASS_NAMES)))),
    y=y_train
)

class_weights = {i: w for i, w in enumerate(class_weights_values)}
print(class_weights)


# =========================================================
# 25. DATA AUGMENTATION
# =========================================================
data_augmentation = keras.Sequential([
    layers.RandomFlip("horizontal"),
    layers.RandomRotation(0.03),
    layers.RandomZoom(0.08),
    layers.RandomContrast(0.08),
    layers.GaussianNoise(0.02),
], name="augmentation")


# =========================================================
# 26. BUILD IMPROVED EFFICIENTNET MODEL
# =========================================================
base_model = keras.applications.EfficientNetB0(
    include_top=False,
    weights="imagenet",
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)

base_model.trainable = False

inputs = keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x = data_augmentation(inputs)
x = base_model(x, training=False)
x = layers.GlobalAveragePooling2D(name="global_average_pooling2d")(x)
x = layers.BatchNormalization(name="batch_normalization")(x)
x = layers.Dense(256, activation="relu", name="dense")(x)
x = layers.Dropout(0.4, name="dropout")(x)
outputs = layers.Dense(len(CLASS_NAMES), activation="softmax", name="dense_1")(x)

model = keras.Model(inputs, outputs)


# =========================================================
# 27. COMPILE MODEL
# =========================================================
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=3e-4, clipnorm=1.0),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

model.summary()


# =========================================================
# 28. CALLBACKS
# =========================================================
callbacks = [
    keras.callbacks.EarlyStopping(
        monitor="val_loss",
        patience=5,
        restore_best_weights=True
    ),
    keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss",
        factor=0.5,
        patience=2,
        min_lr=1e-7,
        verbose=1
    ),
    keras.callbacks.ModelCheckpoint(
        "best_brain_tumor_model.keras",
        monitor="val_loss",
        save_best_only=True,
        verbose=1
    )
]


# =========================================================
# 29. TRAIN STAGE 1
# =========================================================
history1 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=12,
    class_weight=class_weights,
    callbacks=callbacks
)


# =========================================================
# 30. FINE-TUNE STAGE 2
# =========================================================
base_model.trainable = True

for layer in base_model.layers[:-20]:
    layer.trainable = False

model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-6, clipnorm=1.0),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"]
)

history2 = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=10,
    class_weight=class_weights,
    callbacks=callbacks
)


# =========================================================
# 31. PLOT TRAINING HISTORY
# =========================================================
acc = history1.history["accuracy"] + history2.history["accuracy"]
val_acc = history1.history["val_accuracy"] + history2.history["val_accuracy"]
loss = history1.history["loss"] + history2.history["loss"]
val_loss = history1.history["val_loss"] + history2.history["val_loss"]

epochs_range = range(1, len(acc) + 1)

plt.figure(figsize=(14, 5))

plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label="Train Accuracy")
plt.plot(epochs_range, val_acc, label="Validation Accuracy")
plt.title("Training vs Validation Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label="Train Loss")
plt.plot(epochs_range, val_loss, label="Validation Loss")
plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()

plt.tight_layout()
plt.savefig("training_curves.png", dpi=300, bbox_inches="tight")
plt.show()


# =========================================================
# 32. LOAD BEST MODEL AND EVALUATE
# =========================================================
model = keras.models.load_model("best_brain_tumor_model.keras")

test_loss, test_acc = model.evaluate(test_ds)
print("Test Loss:", test_loss)
print("Test Accuracy:", test_acc)


# =========================================================
# 33. PREDICT ON TEST SET
# =========================================================
y_true = []
y_pred = []
y_prob = []

for batch_images, batch_labels in test_ds:
    preds = model.predict(batch_images, verbose=0)
    pred_classes = np.argmax(preds, axis=1)

    y_true.extend(batch_labels.numpy())
    y_pred.extend(pred_classes)
    y_prob.extend(np.max(preds, axis=1))

y_true = np.array(y_true)
y_pred = np.array(y_pred)
y_prob = np.array(y_prob)


# =========================================================
# 34. CLASSIFICATION REPORT
# =========================================================
print(classification_report(
    y_true,
    y_pred,
    target_names=CLASS_NAMES,
    digits=4
))


# =========================================================
# 35. CONFUSION MATRIX
# =========================================================
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(7, 6))
sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=CLASS_NAMES,
    yticklabels=CLASS_NAMES
)
plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")
plt.savefig("confusion_matrix.png", dpi=300, bbox_inches="tight")
plt.show()


# =========================================================
# 36. SHOW SAMPLE PREDICTIONS
# =========================================================
test_paths = test_df["filepath"].tolist()
sample_indices = np.random.choice(len(test_paths), size=min(12, len(test_paths)), replace=False)

plt.figure(figsize=(14, 10))
for i, idx in enumerate(sample_indices):
    img = cv2.imread(test_paths[idx])
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    true_label = CLASS_NAMES[y_true[idx]]
    pred_label = CLASS_NAMES[y_pred[idx]]

    plt.subplot(3, 4, i + 1)
    plt.imshow(img)
    plt.title(f"True: {true_label}\nPred: {pred_label}")
    plt.axis("off")

plt.tight_layout()
plt.show()


# =========================================================
# 37. GRAD-CAM SETUP
# =========================================================
base_model = model.get_layer("efficientnetb0")
last_conv_layer = base_model.get_layer("top_conv")

last_conv_layer_model = tf.keras.Model(
    inputs=base_model.input,
    outputs=last_conv_layer.output
)

classifier_input = tf.keras.Input(shape=last_conv_layer.output.shape[1:])

x = classifier_input
x = model.get_layer("global_average_pooling2d")(x)
x = model.get_layer("batch_normalization")(x)
x = model.get_layer("dense")(x)
x = model.get_layer("dropout")(x)
x = model.get_layer("dense_1")(x)

classifier_model = tf.keras.Model(classifier_input, x)


# =========================================================
# 38. GRAD-CAM FUNCTION
# =========================================================
def make_gradcam_heatmap(img_array, pred_index=None):
    with tf.GradientTape() as tape:
        conv_outputs = last_conv_layer_model(img_array, training=False)
        tape.watch(conv_outputs)

        preds = classifier_model(conv_outputs, training=False)

        if pred_index is None:
            pred_index = tf.argmax(preds[0])

        class_channel = preds[:, pred_index]

    grads = tape.gradient(class_channel, conv_outputs)
    pooled_grads = tf.reduce_mean(grads, axis=(0, 1, 2))

    conv_outputs = conv_outputs[0]
    heatmap = tf.reduce_sum(conv_outputs * pooled_grads, axis=-1)

    heatmap = tf.maximum(heatmap, 0)
    heatmap /= (tf.reduce_max(heatmap) + 1e-8)

    return heatmap.numpy()


# =========================================================
# 39. LOAD SINGLE IMAGE FOR GRAD-CAM
# =========================================================
def load_single_image_for_model(path):
    raw_img = cv2.imread(path)
    raw_img = cv2.cvtColor(raw_img, cv2.COLOR_BGR2RGB)

    proc_img = clahe_rgb(raw_img)
    proc_img = cv2.resize(proc_img, (IMG_SIZE, IMG_SIZE)).astype("float32")
    proc_img = preprocess_input(proc_img)

    return raw_img, proc_img


# =========================================================
# 40. GENERATE GRAD-CAM ON A CONFIDENT CORRECT PREDICTION
# =========================================================
correct_indices = np.where(y_true == y_pred)[0]
best_idx = correct_indices[np.argmax(y_prob[correct_indices])]

sample_path = test_df.iloc[best_idx]["filepath"]
true_label = CLASS_NAMES[y_true[best_idx]]
pred_label = CLASS_NAMES[y_pred[best_idx]]

print("Selected image:", sample_path)
print("True label:", true_label)
print("Predicted label:", pred_label)
print("Confidence:", y_prob[best_idx])

raw_img, proc_img = load_single_image_for_model(sample_path)

img_batch = np.expand_dims(proc_img, axis=0)
heatmap = make_gradcam_heatmap(img_batch, pred_index=y_pred[best_idx])

display_img = cv2.resize(raw_img, (IMG_SIZE, IMG_SIZE))

heatmap = cv2.resize(heatmap, (IMG_SIZE, IMG_SIZE))
heatmap_uint8 = np.uint8(255 * heatmap)

heatmap_color = cv2.applyColorMap(heatmap_uint8, cv2.COLORMAP_JET)
heatmap_color = cv2.cvtColor(heatmap_color, cv2.COLOR_BGR2RGB)

overlay = cv2.addWeighted(display_img, 0.6, heatmap_color, 0.4, 0)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(display_img)
plt.title(f"Original\nTrue: {true_label}")
plt.axis("off")

plt.subplot(1, 3, 2)
plt.imshow(heatmap, cmap="jet")
plt.title("Grad-CAM Heatmap")
plt.axis("off")

plt.subplot(1, 3, 3)
plt.imshow(overlay)
plt.title(f"Overlay\nPredicted: {pred_label}")
plt.axis("off")

plt.tight_layout()
plt.savefig("gradcam_result.png", dpi=300, bbox_inches="tight")
plt.show()


# =========================================================
# 41. SAVE FINAL MODEL
# =========================================================
model.save("final_brain_tumor_efficientnet.keras")
print("Model saved successfully")


# =========================================================
# 42. PREDICT ONE CUSTOM IMAGE
# =========================================================
def predict_one_image(path):
    raw_img = cv2.imread(path)
    raw_img = cv2.cvtColor(raw_img, cv2.COLOR_BGR2RGB)

    img = clahe_rgb(raw_img)
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE)).astype("float32")
    img = preprocess_input(img)

    pred = model.predict(np.expand_dims(img, axis=0), verbose=0)[0]
    pred_idx = np.argmax(pred)

    plt.figure(figsize=(5, 5))
    plt.imshow(cv2.resize(raw_img, (IMG_SIZE, IMG_SIZE)))
    plt.title(f"Predicted: {CLASS_NAMES[pred_idx]}\nConfidence: {pred[pred_idx]:.4f}")
    plt.axis("off")
    plt.show()
```
